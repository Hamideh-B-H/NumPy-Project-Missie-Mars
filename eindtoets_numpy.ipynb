## Deel 1: De Kolonie in Kaart (np.indices & broadcasting)

De Mars-kolonie is gebouwd als een raster van **5 rijen x 6 kolommen** modules. Elke module heeft een positie `(rij, kolom)` in het raster. Het controlecentrum staat op positie `(2, 3)` (midden van het raster).

### Oefening 1.1 — Raster genereren
Gebruik `np.indices` om twee arrays te genereren: een met de **rij-indices** en een met de **kolom-indices** voor het 5x6 raster. Sla ze op in variabelen `rijen` en `kolommen`.

### Oefening 1.2 — Afstand tot het controlecentrum (broadcasting)

Het controlecentrum staat op positie `(2, 3)`. Bereken de **Euclidische afstand** van elke module tot het controlecentrum met broadcasting.

Formule: `sqrt((rij - 2)^2 + (kolom - 3)^2)`

Sla het resultaat op in een variabele `afstanden`. Print de afstanden afgerond op 2 decimalen.

*Tip: je hoeft geen loop te schrijven!*

## Deel 2: Temperatuurmetingen (Broadcasting & Reshape)

De kolonie heeft **4 kassen** die voedsel produceren. Elke kas wordt **3 keer per dag** gemeten (ochtend, middag, avond) gedurende **7 dagen**. De data is opgeslagen in een 3D array met shape `(4, 7, 3)` — (kas, dag, meting).

### Oefening 2.1 — Data genereren

Genereer een 3D array `temperaturen` met shape `(4, 7, 3)` met willekeurige temperaturen (floats) tussen **18.0** en **28.0** graden. Gebruik de `rng` die al gedefinieerd is.

*Tip: `rng.uniform(low, high, size=...)`*

````python
# Oefening 2.1 — Jouw code hier
import numpy as np

rng = np.random.default_rng(2087)
temperaturen = rng.uniform(18.0, 28.0, size=(4, 7, 3))

print(temperaturen)
print(temperaturen.shape)
````

### Oefening 2.2 — Kalibratiecorrectie per kas (broadcasting met reshape)

De sensoren in de 4 kassen zijn niet perfect gekalibreerd. De correctiewaarden per kas zijn:

```python
correctie = np.array([0.5, -0.3, 0.8, -0.1])  # per kas
```

Pas deze correctie toe op **alle** temperatuurmetingen door `correctie` op te tellen bij `temperaturen`. 

**Let op:** de correctie heeft shape `(4,)` en temperaturen heeft shape `(4, 7, 3)`. Je zult `correctie` moeten **reshapen** zodat broadcasting werkt over de juiste as (de eerste as = kas).

Sla het resultaat op in `temp_gekalibreerd`.

````python
correctie = np.array([0.5, -0.3, 0.8, -0.1])

# apply broadcasting by reshaping correction to match (kas, dag, meting)
temp_gekalibreerd = temperaturen + correctie.reshape(4, 1, 1)

# print result
print(temp_gekalibreerd)
print(temp_gekalibreerd.shape)
````

### Oefening 2.3 — Daggewichten toepassen (broadcasting met vector)

De ochtend-, middag- en avondmetingen zijn niet even belangrijk. De gewichten zijn:

```python
gewichten = np.array([0.25, 0.50, 0.25])  # ochtend, middag, avond
```

Bereken voor elke kas en elke dag de **gewogen temperatuur** door `temp_gekalibreerd` element-wise te vermenigvuldigen met `gewichten` en vervolgens te sommeren over de laatste as (`axis=2`).

Het resultaat `temp_gewogen` moet shape `(4, 7)` hebben — de gewogen gemiddelde temperatuur per kas per dag.

````python
gewichten = np.array([0.25, 0.50, 0.25])

# Oefening 2.3 — Jouw code hier
# element-wise multiplication + sum over measurement axis
temp_gewogen = np.sum(temp_gekalibreerd * gewichten, axis=2)

# print result
print(temp_gewogen)
print(temp_gewogen.shape)
````

## Deel 3: Alarmsysteem (Boolean Masks & np.where)

De gewogen temperaturen uit Deel 2 worden nu gebruikt door het alarmsysteem. Er zijn drie zones:

| Zone | Temperatuurbereik | Status |
|------|------------------|--------|
| Veilig | 20.0 - 25.0 | OK |
| Waarschuwing | 18.0-20.0 of 25.0-27.0 | LET OP |
| Alarm | < 18.0 of > 27.0 | ALARM |

### Oefening 3.1 — Veilige metingen filteren

Maak een boolean mask `mask_veilig` die `True` is waar `temp_gewogen` tussen **20.0 en 25.0** ligt (grenzen inclusief).

- Hoeveel metingen zijn veilig? (Gebruik `np.sum` of `.sum()` op de mask)
- Print alle veilige temperatuurwaarden.

````python
# Oefening 3.1 — Jouw code hier
# create boolean mask for safe temperatures (inclusive range)
mask_veilig = (temp_gewogen >= 20.0) & (temp_gewogen <= 25.0)

# count how many safe measurements
aantal_veilig = np.sum(mask_veilig)

# print results
print("Aantal veilige metingen:", aantal_veilig)
print("Veilige temperaturen:", temp_gewogen[mask_veilig])
````

### Oefening 3.2 — Gecombineerde mask: alarm-situaties

Maak een mask `mask_alarm` die `True` is waar de temperatuur **lager dan 18.0 OF hoger dan 27.0** is.

- Print welke temperaturen in alarm-status zijn.
- Gebruik `~` om ook een mask `mask_geen_alarm` te maken en verifieer dat de aantallen kloppen (alarm + geen_alarm = totaal).

````python
# Oefening 3.2 — Jouw code hier

# mask for alarm: too cold OR too hot
mask_alarm = (temp_gewogen < 18.0) | (temp_gewogen > 27.0)

# values in alarm
print("Alarm temperaturen:", temp_gewogen[mask_alarm])

# inverse mask (no alarm)
mask_geen_alarm = ~mask_alarm

# verification of counts
aantal_alarm = np.sum(mask_alarm)
aantal_geen_alarm = np.sum(mask_geen_alarm)
totaal = temp_gewogen.size

print("Aantal alarm:", aantal_alarm)
print("Aantal geen alarm:", aantal_geen_alarm)
print("Totaal:", totaal)

# check correctness
print("Controle klopt:", aantal_alarm + aantal_geen_alarm == totaal)
````

### Oefening 3.3 — Statuslabels toekennen met np.where

Gebruik **geneste `np.where`** om een array `status` te maken met dezelfde shape als `temp_gewogen`, waarin elke meting het label krijgt:
- `'ALARM'` als temp < 18.0 of temp > 27.0
- `'LET OP'` als temp tussen 18.0-20.0 of 25.0-27.0 (exclusief de alarm-grenzen)
- `'OK'` in alle andere gevallen

Print het resultaat.

````python
# Oefening 3.3 — Jouw code hier
status = np.where(
    (temp_gewogen < 18.0) | (temp_gewogen > 27.0),
    "ALARM",
    np.where(
        ((temp_gewogen >= 18.0) & (temp_gewogen < 20.0)) |
        ((temp_gewogen > 25.0) & (temp_gewogen <= 27.0)),
        "LET OP",
        "OK"
    )
)

print(status)
````


## Deel 5: Bemanning Rangschikken (Sorting & Partitioning)

De kolonie heeft 15 bemanningsleden. Elk bemanningslid heeft een **fitnessscore** (0-100) die wekelijks wordt gemeten. De commandant moet beslissingen nemen op basis van deze scores.

### Oefening 5.1 — Scores sorteren

```python

namen = np.array(['Ada', 'Ben', 'Cas', 'Deb', 'Eva', 'Fin', 'Gia', 'Hal', 'Ivy', 'Jan', 'Kim', 'Leo', 'Mia', 'Nik', 'Oli'])
```

1. Sorteer de fitnessscores van laag naar hoog met `np.sort`. Sla op in `scores_sorted`.
2. Gebruik `np.argsort` om de **namen** te sorteren op basis van hun fitnessscore (laag → hoog).
3. Print de top 5 namen met hoogste fitnessscore (hint: sorteer en neem de laatste 5, of keer de volgorde om).

````python
fitness = np.array([78, 92, 45, 88, 67, 55, 91, 73, 82, 60, 95, 38, 71, 84, 50])
namen = np.array(['Ada', 'Ben', 'Cas', 'Deb', 'Eva', 'Fin', 'Gia', 'Hal', 'Ivy', 'Jan', 'Kim', 'Leo', 'Mia', 'Nik', 'Oli'])

# Oefening 5.1 — Jouw code hier
# 1. Sorteer scores
scores_sorted = np.sort(fitness)
print("Gesorteerde scores:", scores_sorted)
# 2. Sorteer namen op basis van score
indices = np.argsort(fitness)
namen_sorted = namen[indices]
print("Namen gesorteerd op score:", namen_sorted)
# 3. Top 5 namen
top5_namen = namen_sorted[-5:]
print("Top 5 hoogste scores:", top5_namen)
````

### Oefening 5.2 — Snel de zwakste 3 vinden (np.argpartition)

De commandant wil snel weten welke **3 bemanningsleden** de laagste fitnessscore hebben, zonder de hele array te sorteren.

2. Print de namen en scores van deze 3 bemanningsleden. (mag met een loop)
3. Gebruik ook `np.partition` met negatief k (`-3`) om de **3 hoogste** scores te vinden.

**Waarom is `np.argpartition` efficiënter dan `np.argsort` voor deze taak?** (Beantwoord als commentaar)

````python
# Oefening 5.2 — Jouw code hier
# 1. indices of 3 lowest scores
indices_laagste = np.argpartition(fitness, 3)[:3]

print("3 laagste scores:")
for i in indices_laagste:
    print(namen[i], fitness[i])

# 2. 3 highest scores using partition
hoogste_scores = np.partition(fitness, -3)[-3:]
print("3 hoogste scores:", hoogste_scores)
````

Vul de array met deze data:

### Oefening 6.2 — Velden opvragen en filteren

Gebruik de structured array `bemanning` om het volgende te doen:

1. Print alle **namen** (veld selectie).
2. Print de **gemiddelde leeftijd** van de bemanning.
3. Filter alle bemanningsleden met een **fitness hoger dan 80** en print hun namen en fitness.
4. Filter alle **Piloten** en print hun namen en aantal missies.

### Oefening 6.3 — Bonus berekenen met np.where

De commandant wil een bonus toekennen:
- Bemanningsleden met **fitness >= 80** krijgen een bonus van **+5 missies**.
- Bemanningsleden met **fitness < 50** krijgen **verplichte rust** (missies worden op **0** gezet).
- Alle anderen blijven ongewijzigd.

1. Maak een kopie van `bemanning` genaamd `bemanning_updated`.
2. Gebruik (geneste) `np.where` om het veld `missies` aan te passen volgens bovenstaande regels.
3. Print voor elk bemanningslid de naam, fitness en het (nieuwe) aantal missies.

````python
# Oefening 6.3 — Jouw code hier

# 1. make a copy
bemanning_updated = bemanning.copy()

# 2. update missions using nested np.where
bemanning_updated['missies'] = np.where(
    bemanning['fitness'] >= 80,
    bemanning['missies'] + 5,
    np.where(
        bemanning['fitness'] < 50,
        0,
        bemanning['missies']
    )
)

# 3. print results
for persoon in bemanning_updated:
    print(persoon['naam'], persoon['fitness'], persoon['missies'])
````

### Oefening 7.2 — Facturatie berekenen

Bereken de factuur in 3 stappen (net als in file 12):

**Stap 1:** Basisfactuur = `zo - ma` (toename over de week).

**Stap 2:** Boete: als `zo > vr` (weekendverbruik stijgt), tel **+20** op bij de factuur. Gebruik `np.where`.

**Stap 3:** Korting: als het verbruik doorheen de werkweek **daalt** (`ma >= di >= wo >= do >= vr`), krijgt de module **15% korting** op de factuur. Gebruik een gecombineerde boolean mask en `np.where`.

Print na elke stap de modulenaam en factuur.

````python
# Oefening 7.2 — Jouw code hier
# Stap 1: Basisfactuur
energie['factuur'] = energie['zo'] - energie['ma']

print("Na stap 1 (basisfactuur):")
for e in energie:
    print(e['module'], e['factuur'])

# Stap 2: Boete weekendverbruik
energie['factuur'] = np.where(
    energie['zo'] > energie['vr'],
    energie['factuur'] + 20,
    energie['factuur']
)

print("\nNa stap 2 (boete):")
for e in energie:
    print(e['module'], e['factuur'])
# Stap 3: Korting dalend weekverbruik
mask_daling = (
    (energie['ma'] >= energie['di']) &
    (energie['di'] >= energie['wo']) &
    (energie['wo'] >= energie['do']) &
    (energie['do'] >= energie['vr'])
)

energie['factuur'] = np.where(
    mask_daling,
    energie['factuur'] * 0.85,  # 15% discount
    energie['factuur']
)

print("\nNa stap 3 (korting):")
for e in energie:
    print(e['module'], e['factuur'])
````

### Oefening 9.3 — Dichtstbijzijnde hulppost (alles samen)

Er is een noodgeval in de kolonie! Het noodsignaal komt van positie `(4, 1)` in het 5x6 raster. Er zijn 3 hulpposten op de volgende posities:

```python
hulpposten = np.array([[0, 0], [2, 5], [3, 2]])
noodpositie = np.array([4, 1])
```

1. Bereken de Euclidische afstand van de `noodpositie` tot elke hulppost met **broadcasting** (niet met een loop!).
2. Gebruik `np.argpartition` (of `np.argmin`) om de **dichtstbijzijnde** hulppost te vinden.
3. Print welke hulppost het dichtst bij is en hoe ver die is.

*Dit combineert broadcasting, afstandsberekening en partitioning/sorting — alles wat je geleerd hebt!*

````python
hulpposten = np.array([[0, 0], [2, 5], [3, 2]])
noodpositie = np.array([4, 1])

# Oefening 9.3 — Jouw code hier

# Data provided for the emergency
hulpposten = np.array([[0, 0], [2, 5], [3, 2]])
noodpositie = np.array([4, 1])

# 1. Calculate the distance using broadcasting
# We subtract the emergency position [4, 1] from all three help station rows at once
differences = hulpposten - noodpositie

# 2. Square the differences and sum them along the columns (axis=1)
# Then take the square root to get the final Euclidean distances
distances = np.sqrt(np.sum(differences**2, axis=1))

# 3. Find the index of the smallest distance (the nearest station)
nearest_index = np.argmin(distances)

# 4. Print results
closest_station = hulpposten[nearest_index]
distance_value = distances[nearest_index]

print(f"The nearest help station is at location: {closest_station}")
print(f"The distance is: {distance_value:.2f} units")
````
